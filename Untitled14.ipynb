```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
def evaluate_and_plot(model, X_test, y_test, title):
    y_pred = model.predict(X_test)

    # 1. Tampilkan Classification Report
    print(f"\n{title}")
    print("-" * 60)
    print(classification_report(y_test, y_pred))

    # 2. Tampilkan Confusion Matrix Heatmap
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
    plt.title(f"Confusion Matrix - {title}")
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```

```python
# Preprocessing minimal (Label Encoding)
df_direct = df.drop('customerID', axis=1)
le = LabelEncoder()
for col in df_direct.select_dtypes(include=['object']).columns:
    df_direct[col] = le.fit_transform(df_direct[col])

# Handle NaN values by dropping rows containing them
df_direct.dropna(inplace=True)

X1 = df_direct.drop('Churn', axis=1)
y1 = df_direct['Churn']
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

# Training 3 Model
m1 = LogisticRegression(max_iter=5000).fit(X_train1, y_train1)
m2 = RandomForestClassifier(random_state=42).fit(X_train1, y_train1)
m3 = VotingClassifier(estimators=[('lr', m1), ('rf', m2)], voting='hard').fit(X_train1, y_train1)

# Evaluasi
evaluate_and_plot(m1, X_test1, y_test1, "Logistic Regression (Scenario 1)")
evaluate_and_plot(m2, X_test1, y_test1, "Random Forest (Scenario 1)")
evaluate_and_plot(m3, X_test1, y_test1, "Voting Classifier (Scenario 1)")
```

```python
scaler = StandardScaler()
X2_scaled = scaler.fit_transform(X1)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2_scaled, y1, test_size=0.2, random_state=42)

m4 = LogisticRegression(max_iter=5000).fit(X_train2, y_train2)
m5 = RandomForestClassifier(random_state=42).fit(X_train2, y_train2)
m6 = VotingClassifier(estimators=[('lr', m4), ('rf', m5)], voting='hard').fit(X_train2, y_train2)

evaluate_and_plot(m4, X_test2, y_test2, "Logistic Regression (Scenario 2)")
evaluate_and_plot(m5, X_test2, y_test2, "Random Forest (Scenario 2)")
evaluate_and_plot(m6, X_test2, y_test2, "Voting Classifier (Scenario 2)")
```

```python
from sklearn.model_selection import GridSearchCV

# Tuning Random Forest
param_grid = {'n_estimators': [50, 100], 'max_depth': [5, 10]}
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3).fit(X_train2, y_train2)

m7 = LogisticRegression(C=0.1, max_iter=5000).fit(X_train2, y_train2)
m8 = grid_rf.best_estimator_
m9 = VotingClassifier(estimators=[('lr', m7), ('rf', m8)], voting='soft').fit(X_train2, y_train2)

evaluate_and_plot(m7, X_test2, y_test2, "Logistic Regression (Scenario 3)")
evaluate_and_plot(m8, X_test2, y_test2, "Random Forest (Scenario 3)")
evaluate_and_plot(m9, X_test2, y_test2, "Voting Classifier (Scenario 3)")

# --- Visualisasi Feature Importance ---
importances = m8.feature_importances_
feat_import_df = pd.DataFrame({'Fitur': X1.columns, 'Skor': importances}).sort_values(by='Skor', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Skor', y='Fitur', data=feat_import_df.head(10))
plt.title('Top 10 Fitur Paling Berpengaruh (Model m8 - Skenario 3)')
plt.show()
```
